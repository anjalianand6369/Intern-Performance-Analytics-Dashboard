```python
# Intern Performance Analytics
## Sprint 1: Data Preparation

# importing required libraries
import pandas as pd
import numpy as np
```

```python
# loading the dataset
df = pd.read_csv(r"C:\Users\91948\Downloads\intern_performance_raw datasets.csv")

# just checking first few rows
df.head()
```

```python
# checking basic info about the data
df.shape
df.columns
df.info()
```

```python
# looking at summary statistics
df.describe()
```

```python
# checking for missing values
df.isnull().sum()
```

```python
# filling missing values using median 
for col in ['communication_score', 'technical_score', 'mentor_feedback_score']:
    df[col] = df[col].fillna(df[col].median())
```

```python
# rechecking missing values
df.isnull().sum()
```

```python
# cleaning text columns
df['gender'] = df['gender'].str.lower().str.strip()
df['department'] = df['department'].str.lower().str.strip()
df['extra_activities'] = df['extra_activities'].str.lower().str.strip()
```

```python
# converting yes/no into numbers
df['extra_activities'] = df['extra_activities'].map({'yes': 1, 'no': 0})
```

```python
# fixing internship duration (handling text + numbers)

df['internship_duration'] = df['internship_duration'].astype(str).str.lower().str.strip()

word_to_num = {
    'one': 1, 'two': 2, 'three': 3, 'four': 4,
    'five': 5, 'six': 6, 'seven': 7, 'eight': 8,
    'nine': 9, 'ten': 10
}

df['internship_duration'] = df['internship_duration'].replace(word_to_num)

df['internship_duration'] = df['internship_duration'].astype(str).str.extract(r'(\d+)')[0]

df['internship_duration'] = pd.to_numeric(df['internship_duration'], errors='coerce')

df['internship_duration'] = df['internship_duration'].fillna(df['internship_duration'].median())

df['internship_duration'] = df['internship_duration'].astype(int)
```

```python
# verifying internship_duration
df['internship_duration'].unique()
df['internship_duration'].isnull().sum()
df['internship_duration'].head()
```

```python
# checking duplicates
df.duplicated().sum()
```

```python
# removing duplicates
df.drop_duplicates(inplace=True)
```

```python
# creating new columns

df['overall_performance'] = (
    df['task_completion_rate'] +
    df['project_score'] +
    df['technical_score']
) / 3

df['soft_skills_score'] = (
    df['communication_score'] +
    df['teamwork_score'] +
    df['initiative_score']
) / 3

df['discipline_score'] = df['attendance_rate'] - df['late_submissions']
```

```python
# final validation
df.isnull().sum()
df.describe()
```

```python
# saving updated cleaned dataset
df.to_csv(r"C:\Users\91948\Downloads\cleaned_intern_dataset.csv", index=False)
```
