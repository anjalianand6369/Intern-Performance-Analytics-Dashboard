```python
## Sprint 2: Exploratory Data Analysis (EDA)

# importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading cleaned dataset
df = pd.read_csv(r"C:\Users\91948\Downloads\cleaned_intern_dataset.csv")

# checking first few rows
df.head()
```

```python
# understanding dataset size and structure
df.shape

# getting statistical summary
df.describe()
```

```python
# checking correlation between numeric variables
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)

plt.title("Correlation Matrix")
plt.show()
```

```python
# checking average performance across departments
df.groupby('department')['overall_performance'].mean().plot(kind='bar')

plt.title("Average Performance by Department")
plt.xlabel("Department")
plt.ylabel("Performance")
plt.show()
```

```python
# checking how attendance affects performance
plt.scatter(df['attendance_rate'], df['overall_performance'])

plt.xlabel("Attendance Rate")
plt.ylabel("Performance")
plt.title("Attendance vs Performance")
plt.show()
```

```python
# checking relationship between technical score and project score
plt.scatter(df['technical_score'], df['project_score'])

plt.xlabel("Technical Score")
plt.ylabel("Project Score")
plt.title("Technical Score vs Project Score")
plt.show()
```

```python
# checking impact of late submissions on performance
plt.scatter(df['late_submissions'], df['overall_performance'])

plt.xlabel("Late Submissions")
plt.ylabel("Performance")
plt.title("Late Submissions vs Performance")
plt.show()
```

```python
# checking distribution of overall performance
df['overall_performance'].plot(kind='hist')

plt.title("Performance Distribution")
plt.xlabel("Performance")
plt.show()
```

```python
# checking distribution of interns across departments
df['department'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title("Department Distribution")
plt.ylabel("")
plt.show()
```

```python
# finding top performing interns
df.sort_values(by='overall_performance', ascending=False).head(10)
```
